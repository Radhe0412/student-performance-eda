## Converting Categorical Values

```python
df["Sex"]=df["Sex"].replace({1:"Female",2:"Male"})
df["Sex"]
```

```python
df["Weekly study hours"]=df["Weekly study hours"].replace({1:"None",2:"<5hrs",3:"6-10hrs",4:"11-20hrs",5:">20hrs"})
df["Weekly study hours"]
```

```python
df["Attendance to classes"]=df["Attendance to classes"].replace({1:"ALways",2:"Sometimes",3:"Never"})
df["Attendance to classes"]
```

```python
df["Taking notes in classes"]=df["Taking notes in classes"].replace({3:"ALways",2:"Sometimes",1:"Never"})
df["Taking notes in classes"]
```

```python
df["Listening in classes"]=df["Listening in classes"].replace({3:"ALways",2:"Sometimes",1:"Never"})
df["Listening in classes"]
```

# Final Conclusion
## This project analyzed factors affecting student academic performance using exploratory data analysis techniques. Attendance to classes, listening in classes, taking notes, and classroom discussion showed positive relationships with grades. Students who actively participated in classroom activities generally performed better academically. Weekly study hours did not show a strong or consistent relationship with grades. Overall, classroom engagement appeared to have a stronger impact on academic performance than study duration alone.
